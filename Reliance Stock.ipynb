# Exploratory data analysis

```python
plt.figure(figsize=(15,8))
df.groupby('Year')['Volume'].mean().plot(kind='bar')
plt.show()
#Maximum number of tradings were done in the year 2005 and least in 2013
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Year'],y=df['Open'],color='red')
plt.show()
#Open value of stock price with respect to year
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Year'],y=df['Close'],color='yellow')
plt.show()
#Close value of stock price with respect to year
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Year'],y=df['High'],color='green')
plt.show()
#High value of stock price with respect to year
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df['Year'],y=df['Low'],color='black')
plt.show()
#Close value of stock price with respect to year
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x=df['High'],y=df['Low'],color='red')
plt.show()
#Visualisation of relation between high and low stock value.
```
